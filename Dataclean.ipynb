```python
import pandas as pd
import re

def clean_name(name):
    if pd.isna(name):
        return None

    name = str(name).strip()

    # Xustomer → Customer
    name = re.sub(r'^Xustomer', 'Customer', name, flags=re.IGNORECASE)

    # ลบ Err / _ERR / Error ที่อยู่ท้ายเท่านั้น
    name = re.sub(r'(?i)(err|_err|error)$', '', name)

    # CUS123 → Customer123 (เฉพาะถ้าตามด้วยเลข)
    name = re.sub(r'^CUS(?=\d)', 'Customer', name, flags=re.IGNORECASE)

    # ลบช่องว่างทั้งหมด
    name = re.sub(r'\s+', '', name)

    # ตัวอักษรให้สวย
    name = name.lower().title()

    # Pattern: Customer0–499
    if re.fullmatch(r'Customer[0-4]?\d{1,2}', name):
        return name
    else:
        return None


# ใช้ฟังก์ชัน
df['name_clean'] = df['name'].apply(clean_name)

# ลบแถวใช้ไม่ได้
df = df[df['name_clean'].notna()].copy()

# แทนที่
df['name'] = df['name_clean']
df = df.drop(columns=['name_clean'])
df['name']
```
