```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.head(19)
```

```python
df['Gender'].unique()
```

```python
df['Gender'] =df['Gender'].astype(str).str.strip().str.lower()

df['Gender'] = df['Gender'].map({
    'm' : 'male'
    ,'f' : 'female'
    ,'male' : 'male'
    ,'female' : 'female'
})
df.head(19)

df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Height(cm)'].isnull().sum()
df['Height(cm)'].fillna(df['Height(cm)'].mean(), inplace=True)
#df['Height(cm)'].isnull().sum()
df['Weight(kg)'].isnull().sum()
df['Weight(kg)'].fillna(df['Weight(kg)'].mean(), inplace=True)
df['BloodPressure'].isnull().sum()
df['BloodPressure'].fillna(df['BloodPressure'].mean(), inplace=True)
df['Cholesterol'].isnull().sum()
df['Cholesterol'].fillna(df['Cholesterol'].mean(), inplace=True)
df = df.dropna(subset=['HeartDisease'])
df[['Smoker',]]
df[['Smoker', 'Exercise', 'Diabetes']].isna().sum()
cols = ['Smoker', 'Exercise', 'Diabetes']  # รายชื่อคอลัมน์ที่ต้องการเติม
df[cols] = df[cols].fillna('Unknown')
cols = ['Smoker', 'Exercise', 'Diabetes']
df[cols] = df[cols].apply(lambda x: x.replace('Unknown', 2))
cols = ['Smoker', 'Exercise', 'Diabetes']

# แปลงข้อความทั้งหมดเป็นตัวพิมพ์เล็กก่อน ป้องกันกรณีมี YES, yes, No, NO
df[cols] = df[cols].apply(lambda x: x.astype(str).str.strip().str.lower())

# แทนค่า Yes → 1, No → 0, Unknown → 2
df[cols] = df[cols].replace({
    'yes': 1,
    'no': 0,
    'unknown': 2
})

# ถ้ายังมีช่องว่างว่างอยู่ (NaN) ให้เติมเป็น 2 เช่นกัน
df[cols] = df[cols].fillna(2)
df.head()
```

```python
df['Exercise'] = df['Exercise'].replace({
    'regular': 1,
    'sometimes': 0,
    2: 2   # ค่าที่ไม่ทราบ (Unknown)
})
df['HeartDisease'] = df['HeartDisease'].replace({
    'Yes': 1,
    'No': 0
})
df['Gender' ] = df['Gender'].replace({
    'male': 1,
    'female': 0
})

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

x = df.drop(columns=['HeartDisease'])
y = df['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
